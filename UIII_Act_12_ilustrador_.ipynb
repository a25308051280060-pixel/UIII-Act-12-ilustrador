#### Ilustrador: Técnica (digital/análoga), tableta gráfica, derechos de autor, galerías de exposición, proyectos editoriales y seguidores.

```python
# --- SISTEMA DE INTELIGENCIA CREATIVA v2.0: CLASE xilustrador ---

class xilustrador:
    """Clase que representa a un artista y gestiona su evolución con IA"""

    def __init__(self, nombre, estilo, experiencia, obras):
        # Atributos de instancia que definen el estado del objeto [3]
        self.nombre = nombre
        self.estilo = estilo
        self.experiencia = experiencia  # Años en la industria
        self.obras = obras              # Cantidad de trabajos realizados

    def analizar_perfil_ia(self):
        """Método de IA que evalúa el rango del artista según sus datos"""
        # Lógica condicional para simular la toma de decisiones [4]
        if self.experiencia >= 10 and self.obras > 50:
            return "MAESTRO SENIOR (Élite)"
        elif self.experiencia >= 3 or self.obras >= 20:
            return "PROFESIONAL (Medio)"
        else:
            return "TALENTO EMERGENTE (Junior)"

    def obtener_fila_tabla(self):
        """Devuelve la información formateada para una tabla alineada [5]"""
        diagnostico = self.analizar_perfil_ia()
        # El formato :<20 asegura que las columnas se mantengan alineadas
        return f"{self.nombre:<18} | {self.estilo:<15} | {self.experiencia:<10} | {self.obras:<10} | {diagnostico}"

    def actualizar_portafolio(self, nuevas_obras):
        """Función interna para modificar el estado del objeto"""
        self.obras = self.obras + nuevas_obras
        print(f"¡IA: Portafolio de {self.nombre} actualizado a {self.obras} obras!")

# --- LÓGICA INTERACTIVA (MENÚ DE 5 OPCIONES) ---

# Lista para almacenar los objetos creados dinámicamente [6]
base_de_datos = []

print("--- AGENCIA DE ILUSTRADORES: SISTEMA IA ACTIVADO ---")

while True:
    print("\n" + "="*90)
    print("      CENTRAL DE MANDO CREATIVO: MENÚ DE 5 OPCIONES")
    print("="*90)
    print("1. Registrar nuevo Ilustrador")
    print("2. Ver Catálogo Completo (Tabla)")
    print("3. Diagnóstico de IA Individual (Búsqueda)")
    print("4. Actualizar Portafolio (Sumar Obras)")
    print("5. Salir del Sistema")
    print("="*90)

    opcion = input("\nSeleccione una acción (1-5): ")

    if opcion == "1":
        # Opción 1: Captura de datos para instanciar el objeto [7]
        print("\n[MODO REGISTRO]")
        nom = input("Nombre: ")
        est = input("Estilo (Digital/Cómic/Arte): ")
        exp = int(input("Años de experiencia: "))
        obr = int(input("Obras actuales: "))

        # Se crea el objeto xilustrador y se añade a la lista
        nuevo = xilustrador(nom, est, exp, obr)
        base_de_datos.append(nuevo)
        print(f"¡IA: El ilustrador '{nom}' ha sido ingresado al sistema!")

    elif opcion == "2":
        # Opción 2: Mostrar listado en formato de tabla alineada
        if not base_de_datos:
            print("\nIA Alerta: No hay registros para mostrar.")
        else:
            print("\n" + "-"*95)
            print(f"{'NOMBRE':<18} | {'ESTILO':<15} | {'EXP (años)':<10} | {'OBRAS':<10} | {'RANGO IA'}")
            print("-" * 95)
            for artista in base_de_datos:
                print(artista.obtener_fila_tabla())
            print("-" * 95)

    elif opcion == "3":
        # Opción 3: Búsqueda y análisis detallado de un objeto específico
        if not base_de_datos:
            print("\nIA: Base de datos vacía.")
        else:
            busqueda = input("Ingrese el nombre del artista para diagnosticar: ")
            encontrado = False
            for artista in base_de_datos:
                if artista.nombre == busqueda:
                    rango = artista.analizar_perfil_ia()
                    print(f"\n>>> ANÁLISIS DE IA PARA {artista.nombre}:")
                    print(f"Estado Profesional: {rango}")
                    print(f"Sugerencia: Su estilo '{artista.estilo}' tiene alta demanda.")
                    encontrado = True
            if not encontrado:
                print("IA: Artista no localizado.")

    elif opcion == "4":
        # Opción 4: Interactuar con un objeto para modificar sus atributos
        if not base_de_datos:
            print("\nIA: No hay perfiles para actualizar.")
        else:
            busqueda = input("¿A qué artista desea sumarle obras nuevas?: ")
            for artista in base_de_datos:
                if artista.nombre == busqueda:
                    cantidad = int(input(f"¿Cuántas obras nuevas terminó {artista.nombre}?: "))
                    artista.actualizar_portafolio(cantidad)

    elif opcion == "5":
        # Opción 5: Terminación del bucle [8]
        print("\nIA: Finalizando sesión. ¡Mucha creatividad y éxito!")
        break

    else:
        print("\n[!] Opción inválida. Intente de nuevo.")

print("\n--- PROGRAMA FINALIZADO ---")
print("programa terminado por Ambar Fierro NL = 19")
```
